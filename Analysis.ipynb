```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

#Decided to play around with the .csv I used for my project. I found out that I'd interpreted some data incorrectly
#in my project, but nonetheless, it was cool to mess with the data in Jupyter. I used what you provided as a base. 

data = pd.read_csv('cafosdata.csv')
```

```python
data[:1]
```

```python
data.sort_values('BROILERS', ascending=False)[:25]
```

```python
data["total_count"] = data["BEEF"] + data["DAIRY"] + data["BROILERS"] + data["LAYERS"] + data["PULLETS"] + data["TURKEYS"]

data.sort_values('total_count', ascending=False)
```

```python

# Highlight Pinellas with facetgrid
g = sns.FacetGrid(data, size=7)

g.map(plt.scatter, "SUB_WATERS", "total_count", alpha=.8)
```

```python
#there's a giant outlier screwing up my axis scale, so I took decided to only include data points with values less 
#than 80,000 in the SUB_WATERS category

smaller_data = data[data.SUB_WATERS < 80000]

g = sns.FacetGrid(smaller_data, size=7)

g.map(plt.scatter, "SUB_WATERS", "total_count", alpha=.8)
```

```python
#Turns out, AU is measured differently than just number of animals, so the total count here was wrong. We need the PF_TOTALAU field.

smaller_data = data[data.SUB_WATERS < 80000]

g = sns.FacetGrid(smaller_data, size=7)

g.map(plt.scatter, "SUB_WATERS", "PF_TOTALAU", alpha=.8)
```
